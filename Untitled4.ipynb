# **EDA & FEATURE ENGINEERING**

Наша цель в рамках поставленной задачи построить скоринговую модель, которая
позволит максимизировать совокупную экономическую прибыль (в процентах от суммы
выдачи). В описании к задаче дана формула для рассчета экономической прибыли:

*   Loss = Unpaid * (EL, expected loss)
*   InvestorProfit = Earned — Loss
*   Profit = InvestorProfit + Commission
*   Profit% = Profit / LoanIssued

где:
*   Unpaid - невыплаченная сумма с процентами;
*   EL - ожидаемый убыток;
*   Earned ınterst - полученные инвестором проценты.

Несмотря на то, что обычно рекомендуется исследовать тренировочную часть отдельно от тестовой, в данном случае у нас нет таргета для модели - нам его предстоит создать на основе рассчетов экономической выгоды. По-этому часть исследований и рассчетов новых признаков будет проводиться на всем датасете.

Исходя из заданной формулы, вероятно что **Commission** - получаемая сервисом плата за организацию выдачи займа в процетах от его суммы. Однако, мы по прежнему не знаем наверняка как именно она рассчитывается и взымается. Попробуем это выяснить:

```python
print(len([i for i in df['Loan issued'] if i%10000 == 0]))
print(len([i for i in df['Loan issued'] if i%10000 != 0]))
```

Мы видим что большинство кредитов - округленные суммы. Интуитивное предположение состоит в том, что комиссия не включена в эту сумму, а рассчитывается от нее, так как большинство людей склонно подавать заявки на круглые суммы. Не круглые суммы займов можно объяснить многими причинами (такими как привязка конечных рассчетов заемщиков с контрагентами к курсу иностранной валюты или точная сумма госкантракта), но для окончательного вывода у нас недостаточно информации.

```python
df_repaid = df[df['Unpaid,  full amount'] == 0]
print('Количество заемщиков, погасивших займ:', len(df_repaid))
print('Минимальная сумма, выплаченная инвестору в процентах от выданной суммы займа:',
        ((df_repaid['Earned interest'] / df_repaid['Loan issued']).min()).round(2))
print('Максимальая сумма, выплаченная инвестору в процентах от выданной суммы займа:',
        ((df_repaid['Earned interest'] / df_repaid['Loan issued']).max()).round(2))
print('Медианная сумма, выплаченая инвестору в процентах от выданной суммы займа:',
        ((df_repaid['Earned interest'] / df_repaid['Loan issued']).median()).round(2))
print('Средняя сумма, выплаченая инвестору в процентах от выданной суммы займа:',
        ((df_repaid['Earned interest'] / df_repaid['Loan issued']).mean()).round(2))
```

```python
print('СТАВКИ КОМИССИИ СРЕДИ ПОГАСИВШИХ ЗАЙМ, %')
print('Минимальная:', df_repaid['Comission, %'].min())
print('Максимальная:', df_repaid['Comission, %'].max())
print('Медиана:', df_repaid['Comission, %'].median())
print('Средняя:', df_repaid['Comission, %'].mean().round(2))
```

Так как половина от тех, кто погасил займ заплатили инвестору менее 4% при максимальном показателе в 15% дает возможность предположить что они погасили займ досрочно. При этом медианная ставка Comission не сильно отличается от максимального. Из этого мы можем сделать вывод, что комиссия включена в регулярные платежи и ее итоговая сумма в рублях зависит не только от суммы займа, но и от срока, в течении которого заемщик погасил кредит. В противном случае заемщику было бы экономически не выгодно досрочно погашать займ.

```python
# Приводим проценты в дробный формат:
df[['Comission, %', 'EL']] = df[['Comission, %', 'EL']].astype('float64')/100
```

```python
# Вычисляем экономическую прибыль на оснавании формулы:
df['loss'] = df['Unpaid,  full amount'] * df['EL']
df['investors_profit'] = df['Earned interest'] - df['loss']
# Размер комиссии уменьшаем на % EL:
df['profit'] = df['investors_profit'] + df['Comission, %'] * df['EL'] * df['Loan issued']
df['profit_%'] = df['profit'] / df['Loan issued']
```

Несмотря на то, что нулевой доход не ухудшает наш ключевой показатель, мы считаем что такие кредиты не должны выдаваться, так как существуют накладные расходы на выдачу и обслуживание кредита, хоть они и не указаны в предоставленном датасете.

```python
df['target'] = np.nan
for i in range(len(df)):
    if df['profit_%'][i] <= 0:
        df['target'][i] = 0
    else:
        df['target'][i] = 1
```

# **DATA PREPROCESSING & MODEL BUILDING**

Подготовим данные для обучения. В качестве классификатора мы будем использовать CatBoost.

```python
# Разбиваем датасет на тренировочную и тестовую части:
FEATURES = [i for i in df.columns if i not in [
    'Earned interest', 'Unpaid,  full amount','loss',
    'investors_profit', 'profit', 'profit_%', 'target']]

CAT_FEATURES = list(df.columns[6:28])

NUM_FEATURES = [i for i in FEATURES if i not in CAT_FEATURES]

df = df[df['Unpaid,  full amount'] != 0]  # Удаляем аномальные строки с EL 100% и строки с новыми заемщиками, не успевшими сделать ни одного платежа
df = df[df['EL'] != 0].reset_index(drop=True)

X = df[FEATURES]
y = df['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=43, stratify=y)

print(y_train.value_counts(), y_test.value_counts())
```

```python
# Определяем веса таргета:
neg, pos = np.bincount(y_train)

total = neg + pos

weight_for_0 = (1 / neg) * (total / 2.0)
weight_for_1 = (1 / pos) * (total / 2.0)

class_weight = {0: weight_for_0, 1: weight_for_1}

print('Weight for class 0: {:.2f}'.format(weight_for_0))
print('Weight for class 1: {:.2f}'.format(weight_for_1))
```

Обучим базовую модель и отберем признаки для дальнейшего обучения:

```python
# Задаем гиперпараметры, нициируем модель для отбора признаков:
cb_classifier = CatBoostClassifier(
                n_estimators=1000,
                eval_metric='AUC',
                random_state=42,
                verbose=200,
                learning_rate = 0.2,
                use_best_model=True,
                early_stopping_rounds=400,
                cat_features=CAT_FEATURES)

base_model = cb_classifier.fit(X_train, y_train, eval_set=(X_test, y_test))
```

```python
# Прогнозируем:
base_predict = base_model.predict(X_test)
target_names = ['class 0', 'class 1']
print(classification_report(base_predict, y_test, target_names=target_names))
print(confusion_matrix(y_test, base_predict , labels=[1,0]))
```

```python
# Визуализируем важность признаков:
feature_importance = base_model.get_feature_importance(type='FeatureImportance')
feature_names = base_model.feature_names_

feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': feature_importance})

feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df)
plt.show()
```

```python
# Отбираем ТОП признаков:
TOP_FEATURES = list(feature_importance_df.sort_values(by='Importance', ascending=False).head(15)['Feature'])
```

Совместим кросс-валидацию и подбор гиперпараметров модели:

```python
# Настраиваем кросс-валидацию:
kf = KFold(n_splits=10, shuffle=True, random_state=42)

ID_LIST = X_train.index

oof = pd.DataFrame(data=np.zeros((len(ID_LIST),1)), index=ID_LIST)

X_train_fold = X_train[TOP_FEATURES].join(y_train)
models = {}

label = y_train
```

```python
# Функция построения словаря моделей с разными гиперпараметрами:
def model_training():
    models_dict = {}

    learning_rate = [0.3, 0.5, None]
    max_depths = [1, 3, None]
    class_weights = [class_weight, None]
    CAT_FEATURES = [i for i in TOP_FEATURES if i not in NUM_FEATURES]

    for rate in learning_rate:
        for debth in max_depths:
            for weight in class_weights:
                cb_classifier = CatBoostClassifier(
                                n_estimators=2000,
                                eval_metric='AUC',
                                random_state=42,
                                verbose=200,
                                learning_rate = rate,
                                class_weights=weight,
                                max_depth=debth,
                                use_best_model=True,
                                early_stopping_rounds=400,
                                cat_features=CAT_FEATURES)


                models_dict[f"cat_boost_model_{rate}_{debth}_{weight}"] = cb_classifier

    return models_dict
```

```python
# Создаем словарь моделей:
models_dict = model_training()
len(models_dict)
```

```python
# Тренируем модели, сохраняем обученные в словарь:
for i, (train_index, valid_index) in enumerate(kf.split(X=X_train_fold)):

        train_df = X_train_fold.iloc[train_index]
        valid_df = X_train_fold.iloc[valid_index]
        valid_ids = valid_df.index.values

        X_train = train_df.drop('target', axis=1)
        X_val = valid_df.drop('target', axis=1)
        y_train = train_df['target']
        y_val = valid_df['target']

        for key in models_dict:
            print(f"fold_{i+1}_{key}")

            models_dict[key].fit(X_train, y_train, eval_set=(X_test[TOP_FEATURES], y_test))

            models[f"fold_{i+1}_{key}"] = models_dict[key]

            y_pred = models_dict[key].predict(X_val)
            oof.loc[valid_ids, 0] = y_pred.flatten()
            target_names = ['class 0', 'class 1']
```

```python
# Просмотр лучшей модели:
best_model = None
best_score = 0

for model_name, model in models.items():
    score = model.get_best_score()
    validation_AUC = score['validation']['AUC']

    if validation_AUC > best_score:
        best_model = model
        best_score = validation_AUC

print(best_model)
print(best_score)
```

```python
# Тестирование лучшей модели:
best_model_predict = best_model.predict(X_test[TOP_FEATURES])
target_names = ['class 0', 'class 1']
print(classification_report(best_model_predict, y_test, target_names=target_names))
print(confusion_matrix(y_test, best_model_predict , labels=[1,0]))
```

Лучшая модель не показала удовлетворительного результата. Попробуем усреднить прогнозы всех обученных на кроссвалидации моделей:

```python
# Создание датасета прогнозов всех обученных моделей:
predict_features = []
predict_features = pd.DataFrame(predict_features)
for key in models:
    predictions = models[key].predict_proba(X_test[TOP_FEATURES])
    predictions = pd.DataFrame(predictions)
    predict_features['{}{}'.format(key, '_class_0')] = predictions[0]
    predict_features['{}{}'.format(key, '_class_1')] = predictions[1]

drop_indexes = list(range(0, len(predict_features.columns), 2))

predict_features_short = predict_features.drop(predict_features.columns[drop_indexes], axis=1)
predict_features_short['target'] = list(y_test)
```

```python
# Усреднение прогнозов:
predict_features_short['pred'] = (predict_features_short.drop('target', axis=1)).sum(axis=1)/predict_features_short.shape[1]
```

```python
# Классификация:
predict_features_short['total_pred'] = np.nan
for i in range(len(predict_features_short['pred'])):

    if predict_features_short['pred'][i] >= 0.5:
        predict_features_short['total_pred'][i] = 1
    else:
        predict_features_short['total_pred'] [i] = 0
```

```python
# Результаты усреднения:
print(classification_report(predict_features_short['target'], predict_features_short['total_pred'], target_names=target_names))
print(confusion_matrix(predict_features_short['target'], predict_features_short['total_pred'] , labels=[1,0]))
```

Усреднение показало себя лучше, но все же недостаточно хорошо. Перейдем к созданию следующего элемента ансамбля - логистической регрессии:

```python
# Задаем гиперпараметры для логистической регрессии:
final_classifier = LogisticRegression(
                    penalty='l2',
                    C=0.5,
                    random_state=42,
                    solver='liblinear',
                    verbose=True,
                    class_weight=class_weight)
```

```python
len(predict_features_short)
```

```python
# Делим на тренировочную и тестовую выборки:
X_COLUMNS = [i for i in predict_features_short.columns if i not in ['target', 'pred', 'total_pred']]

X = predict_features_short[X_COLUMNS]
y = predict_features_short['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=43, stratify=y)
```

```python
# Обучаем финальную модель:
final_model = final_classifier.fit(X_train, y_train)
```

```python
# Классификация:

logreg_pred = final_model.predict_proba(X_test)
logreg_predictions = []
for i in  range(len(logreg_pred)):
    if logreg_pred[i][0] >=0.49:
        logreg_predictions.append(0)
    else:
        logreg_predictions.append(1)
target_names = ['class 0', 'class 1']
print(classification_report(y_test, logreg_predictions, target_names=target_names))
print(confusion_matrix(y_test, logreg_predictions , labels=[1,0]))
```

```python
# Проверка LogLoss:
print('loss without calibration:{0}'.format(log_loss(y_test, logreg_predictions, eps=1e-15, normalize=True)))
```

Попробуем уменьшить ошибку:

```python
# Калибровка LogLoss:
calibrated_logreg = CalibratedClassifierCV(final_model, method='isotonic', cv=5)
calibrated_logreg.fit(X_train, y_train)
print('loss with calibration:{0}'.format(log_loss(y_test, calibrated_logreg.predict_proba(X_test), eps=1e-15, normalize=True)))
```

```python
# Тест откалиброванной модели:
calibrated_logreg_pred = calibrated_logreg.predict_proba(X_test)
calibrated_logreg_predictions = []
for i in  range(len(calibrated_logreg_pred)):
    if calibrated_logreg_pred[i][0] > 0.48:
        calibrated_logreg_predictions.append(0)
    else:
        calibrated_logreg_predictions.append(1)
target_names = ['class 0', 'class 1']
print(classification_report(y_test, calibrated_logreg_predictions, target_names=target_names))
print(confusion_matrix(y_test, calibrated_logreg_predictions, labels=[1,0]))
```

# **ИТОГИ И РЕКОМЕДАЦИИ**

Откалиброванная модель получилась чуть более чувствительной к негативным факторам, что позволит лучше отфильтровывать пограничные факторы (при отсечке 0.5 модель показывает 2 ложно отрицательных и 1 ложноположитеьный результаты). При подборе точки отсечения класса, акцент делался на минимизации ложноположительных прогнозов, так как их стоимость для бизнеса выше.

Вместе с тем, хотелось бы дать несколько рекомендаций:
1. Необходимо дополнить данные датой выдачи кредита, сроком выдачи, % инвестора, дисконтированием, количеством дней просрочки, размерами штрафов и пени.  
2. Необходимо добавить в описание формулу начисления процентов и комиссии.
3. Скорректировать формулу:

*   Loss = adjusted_outstanding_loan * EL
*   InvestorProfit = Earned interest — Loss + investors_unpaid_settlement_balance * (1-EL)
*   Profit = InvestorProfit + commission paid + estimated unpaid commission * (1-EL)
*   Profit% = Profit / LoanIssued
где:
