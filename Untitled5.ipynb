```python
# Необходимые импорты:

import pandas as pd
import numpy as np
import statistics as stat
import matplotlib.pyplot as plt
import warnings

from google.colab import auth
import gspread
from google.auth import default
```

```python
# Настройки ноутбука:
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.float_format', '{:.2f}'.format)
plt.style.use('dark_background')
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
# Путь к файлу, преобразование в датафрейм:
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)
worksheet = gc.open('ТЗ аналитик данных').sheet1
rows = worksheet.get_all_values()
df = pd.DataFrame.from_records(rows)
df.columns = df.iloc[1]
df = df.iloc[2:].reset_index(drop=True)
```

```python
# Удаляем лишние символы в строках:
bad_symbols = ['%', '₽', ' ', '\xa0']
for i in bad_symbols:
    df[df.columns]= df[df.columns].replace(i,'', regex=True)
df[['Comission, %', 'EL']]  = df[['Comission, %', 'EL']].replace(',', '.', regex=True)
# Удаляем пустой столбец:
df.drop(df.columns[6], axis=1, inplace=True)
```

```python
# Преобразование типов данных:
FLOAT_COLUMNS = ['Comission, %', 'EL']
INT_COLUMNS = [i for i in df.columns if i not in FLOAT_COLUMNS]
df[FLOAT_COLUMNS] = df[FLOAT_COLUMNS].astype('float64')
for column in INT_COLUMNS:
    for i in df[column]:
        i = int(i.strip())
df[INT_COLUMNS] = df[INT_COLUMNS].astype('int64')

df.dtypes.value_counts()
```

```python
pivot_table = pd.pivot_table(df, values='Loan issued', index='Comission, %', columns='Рейтинг', aggfunc=np.mean)
pivot_table = pivot_table.fillna(0)
```
